Integrale di $f$ da $a$ fino a $b$ con $n$ intervalli

```python
def integrale(f, a, b, n):
    A=0
    sum_y=0
    x = a
    dx=(b - a) / n
    for i in range(n):
        sum_y= sum_y + f(x)
        ##print(i, A)
        x = x+dx
        A=sum_y*dx
    return A
```

```python
Area = integrale(f, 0, np.pi/2, 10000000)
print('Area sotto la curva = ', Area)
```
