Integrale di $f$ da $a$ fino a $b$ con $n$ intervalli

```python
def integrale(f, a, b, n):
    A=0
    x = a
    dx=(b - a) / n
    for i in range(n):
        A= A + dx * f(x)
        print(i, A)
        x = x+dx
    return A
```

```python
integrale(f, 0, np.pi/2, 4)
```
